## 檔案基本操作

```python
# open file
data = open('example.csv')
data
```

```python
# open file (utf-8)
# readlines()：讀取整個文件所有行，保存於list內
data = open('example.csv', encoding='utf-8')
data.readlines()
```

```python
# open file (utf-8)
# read([size])：從文件當前位置起讀取size個字元數量，若無參數，讀取至文件結束為止
data = open('example.csv', encoding='utf-8')
data.read()
```

```python
# open file (utf-8)
# readline()：讀取一行內容
data = open('example.csv', encoding='utf-8')
data.readline()
```

```python
# file is not in my document, so built a new one automatically.
# write the words in the file and close.
output = open('0610.txt', mode='w', encoding='utf-8')
output.write('Hello\n')
output.close()
```
