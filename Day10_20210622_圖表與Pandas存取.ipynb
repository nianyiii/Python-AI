## 地端操作：開放資料平台下載json檔案(結合pandas)

    資料轉換：json->dict(json.load)->pandas->excel->excel_power map產生影像圖

```python
# 自開放資料平台下載json檔案實作

with open('youbike_immediate.json', 'r', encoding = 'utf-8') as f:
    data = f.read()
data[ :300] # 前300個字元
```

```python
# 承上，轉成list底下的dict

import json

with open('youbike_immediate.json', 'r', encoding = 'utf-8') as f:
    data = json.load(f)
data[ :3] # 前3筆資料
```

```python
# numpy：0-rank, 1-rank, 2-rank, 3-rank
```

```python
# pandas = 2-rank numpy + header = row + column

import pandas as pd

pd.DataFrame(data) # --> 將上述data放進去，產生標題 + 筆數(index) + 欄位(columns)
```

```python
# 再匯出excel格式

import pandas as pd

pd.DataFrame(data).to_excel('./bike.xlsx', index = False)
```

## 雲端操作：開放資料平台爬蟲json檔案(結合pandas)   
    資料轉換：json->dict(json.loads)->pandas

```python
import requests as r
content = r.get('https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json')
content.text
```

```python
# 使用json.loads將data從記憶體轉至json

import json

data = json.loads(content.text)
data
```

```python
import pandas as pd

pd.DataFrame(data)
```
